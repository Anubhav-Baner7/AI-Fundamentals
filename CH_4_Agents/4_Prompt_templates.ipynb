### Chains

```python
# Step 1: Create a prompt template
my_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a GenZ Poet"),
        ("user", "Write a poem about {topic}")
    ]
)

# Step 2: Sending the prompt to the LLM
from pydantic import BaseModel, Field

class llm_response(BaseModel):
    query: str = Field(description="The question that was asked to the LLM")
    answer: str = Field(description="The answer generated by the LLM")
    total_tokens: int = Field(description="The total number of tokens used in the response")

llm_structured = llm.with_structured_output(llm_response)

# Step 3: Parsing the output
from langchain_core.output_parsers import PydanticOutputParser
output_parser = PydanticOutputParser(pydantic_object=llm_response)

#Final Step: Putting it all together

my_chain = my_template | llm_structured 
```

```python
my_chain.invoke({"topic": "the beauty of seas"})
```
