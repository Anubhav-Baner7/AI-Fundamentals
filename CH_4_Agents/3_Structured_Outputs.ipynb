```python
from langchain_openai import ChatOpenAI 
import os
from dotenv import load_dotenv
load_dotenv()
llm = ChatOpenAI(model ="gpt-5-mini") 
```

```python
response = llm.invoke("What is the capital of France?. Generate the response in this format: Query: <the question> \n Answer: <the answer>")
print(response.content)
response
```

### Pydantic

```python
from pydantic import BaseModel, Field

class llm_response(BaseModel):
    query: str = Field(description="The question that was asked to the LLM")
    answer: str = Field(description="The answer generated by the LLM")
    total_tokens: int = Field(description="The total number of tokens used in the response")
```

### Response Validation (Data Validation)

```python
obj = llm_response(query="What is the capital of France?", answer="The capital of France is Paris.", total_tokens=10)
print(obj.query)
print(obj.answer)
print(obj.total_tokens)
obj
```

### Structured Output with LLMs

```python
llm_structured = llm.with_structured_output(llm_response)
response_structured = llm_structured.invoke("What is the capital of France?")
print(response_structured)
```

```python
response_structured.model_dump()
```

```python
response_dict = response_structured.model_dump()
print(response_dict['query'])
print(response_dict['answer'])
print(response_dict['total_tokens'])
```
